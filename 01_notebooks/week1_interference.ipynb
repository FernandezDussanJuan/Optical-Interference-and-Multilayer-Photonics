# Interference of two plane waves
## 1. Monochormatic electromagnetic wave
An electromagnetic light wave can be described as a harmonic electromagnetic wave.
For a plane wave propagating along the $x$ direction the electic field can be written as:

$E(x,t) = E_0 \cos(kx - \omega t + \phi)$


here:
- $E_0$ is the amplitude of the electric field,
- $k = \frac{2\pi}{\lambda}$ is the wave number,
- $\lambda$ is the wavelength, represents the spatial period of the wave,
- $\omega = 2\pi f$ is the angular frequency,
- $\phi$ is the initial phase.

In classical optics, the directly observable quantity is not the electric field itself, but the **intensity**, which is proportional to the time-averaged value of the squared electric field $E^2$.

## 2. Principle of Superposition
Wave optics is bases on the **principle of Superposition**.
As discussed in Hectch, when two or more electromagnetic waves overlap in space, the resulting electric field is given by the sum of the individual fields:

$E_{\text{total}} = E_1 + E_2$

This principle holds in linear media such as air, vacuum and common optical material under normal conditions.

## 3. Interference of two Coherent waves
Interference is defined as the spatial redistribution of intensity produced by the superpositions of two coherent waves.
Two waves are said to be **coherent** if they have the same frequency and maintain constant the phase differences.

We consider two plane waves with of equal amplitude:

$E_1(x,t) = E_0 \cos(kx - \omega t)$

$E_2(x,t) = E_0 \cos(kx - \omega t + \phi)$

The resultant electric field is:

$E_{\text{total}}(x,t) = E_1(x,t) + E_2(x,t)$

The observed intensity is proportional to the time-averaged square of the total field:

$I \propto \langle E_{\text{total}}^2 \rangle$

After time averaging, one obtains:

$I \propto 2E_0^2 \left( 1 + \cos \phi \right)$

The results shows the explicit dependency with the phase difference between the waves, leading to constructive and destructive interference.

## 4. Spatial Dependence of the phase
If the pase depends on position, it can be written as:

$\phi(x) = kx$

As a consequence the intensity becomes position-dependent:

$I(x) \propto \cos^2\left(\frac{kx}{2}\right)$

The spatial modulation of the intensity gives rise to an interference pattern, which is the main phenomenon studied in this project.

### 5. Computacional approach
In this project, the analytical expressions derived from classical wave optics are implemented numerically using Python.  
The electric fields are calculated directly from the theoretical formulas, and the resulting intensity patterns are visualized to study their dependence on physical parameters such as wavelength and phase difference.

This approach follows the classical treatment presented in Hecht, *Optics*, and provides a computational perspective on fundamental interference phenomena.

## Physical Parameters
The simulations are performed using a wavelenght in the visible range $\lambda = 500\ nm$. The spatial domain is chosen clearly t solve the interference fringes and facilitate the observation of the interference pattern.

```python
import os

folder_path = "../figures/week 1"
os.makedirs(folder_path, exist_ok=True)

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Physical parameters
lambda_ = 500e-9  #wavelenght [m]
k = 2*np.pi / lambda_  #
phi = 0  #phase difference

#spatial domain
x = np.linspace(-2e-6,2e-6,2000)
```

## Numerical simulation
We then define the electric field of the two plane waves with an initially zero phase difference. The resulting electric field is the one explained above as the superposition of the electric fields of each wave.

```python
E1 = np.cos(k*x)
E2 = np.cos(k*x + phi)
E_total = E1 + E2  # resultant electric field
I= E_total**2   #Intensity 
```

```python
plt.plot(x*1e6, I)
plt.xlabel("Positon ($\mu$m)") 
plt.ylabel("Intensity (arb.units)")
plt.title("Interference pattern of two plane waves")
plt.savefig(f"{folder_path}/interference_lambda_500nm.png", dpi=300)
plt.show()
```

## Effect of wavelenght and phas difference
To analyze the interference behavior of plane waves, we studied their dependencies on wavelength and phase difference as two isolated cases.

```python
wavelenghts = [400e-9, 500e-9, 600e-9]  # different selected wavelengths

plt.figure()
for lambda_ in wavelenghts:
    k = 2*np.pi / lambda_
    E1 = np.cos(k*x)
    E2 = np.cos(k*x)  # phase difference equal to zero
    I = (E1+E2)**2
    plt.plot(x*1e6, I, label= f"$lambda$={int(lambda_*1e9)} nm")

plt.xlabel("Positon ($\mu$m)") 
plt.ylabel("Intensity (arb.units)")
plt.legend()
plt.title("Effect of wavelenght on Interference pattern")
plt.savefig(f"{folder_path}/Interference_wavelenght.png", dpi=300)
plt.show()
```

```python
phases = [0, np.pi/4, np.pi/2]

plt.figure()
for phi in phases:
    E1 = np.cos(k*x)
    E2 = np.cos(k*x + phi) # phase difference alternated
    I = (E1+E2)**2
    plt.plot(x*1e6, I, label= f"$\phi$={phi:.2f} rad")

plt.xlabel("Positon ($\mu$m)") 
plt.ylabel("Intensity (arb.units)")
plt.legend()
plt.title("Effect of phase difference on Interference pattern")
plt.savefig(f"{folder_path}/Interference_phasedifference.png", dpi=300)
plt.show()
```

## Results and discussion
The result show that increasing the wavelength leads to a larger spacing between interference fringes, in agreement with the theorical dependence on the wave number $k =\dfrac{2\pi}{\lambda}$.

Changing the phase difference produces a spatial shift of the interference pattern without altering the fringe spacing. This behavior is consistent wiht the theorical expression for the intensity distribution.

The results confirms that interference is direct consequence of the linear superposition principle. The spatial modulation of intensity arises from phase difference rather than changes in amplitude, highlighting the wave nature of light.

Interestingly, the mathematical stucture of the interference pattern is formally identical to that obtained in quantum mechanics fowr matter waves,
illustrating the deep connection between classical and quantum wave phenomena.
